```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import pylab
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
training_annotations = '/root/waymo/split_dataset/annotations/valid_annotations.json'
```

```python
coco=COCO(training_annotations)
```

```python
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
print('COCO categories: \n{}\n'.format(' '.join(nms)))

nms = set([cat['supercategory'] for cat in cats])
print('COCO supercategories: \n{}'.format(' '.join(nms)))
```

```python
catIds = coco.getCatIds(catNms=['TYPE_VEHICLE','TYPE_PEDESTRIAN','TYPE_CYCLIST']);
imgIds = coco.getImgIds();
```

```python
catIds
```

```python
len(imgIds)
```

```python
import os
for imgId in range(1,len(imgIds),50):
    print(imgId)
    I = cv2.imread(os.path.join('/root/waymo/split_dataset/valid/','{}.jpg'.format(imgId)))
    annIds = coco.getAnnIds(imgIds=imgId, catIds=catIds, iscrowd=None)
    anns = coco.loadAnns(annIds)
    color_p = (255, 0, 0)
    color_v = (0,255,0)
    for annotation in anns:
        bbox = annotation['bbox']
        if annotation['category_id']==2:
            color = color_v
        else:
            color = color_p
        cv2.rectangle(I,(int(bbox[0]),int(bbox[1])),(int(bbox[0]+bbox[2]),int(bbox[1]+bbox[3])),color=color,thickness=2)
    cv2.imwrite(os.path.join('valid_annotation_split_dataset','{}.jpg'.format(imgId)),I)
```
