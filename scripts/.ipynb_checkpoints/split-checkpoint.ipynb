```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import pylab
pylab.rcParams['figure.figsize'] = (8.0, 10.0)
```

```python
training_annotations = '/root/waymo/data_2d_od/annotations_small/train_annotations.json'
```

```python
coco=COCO(training_annotations)
```

```python
cats = coco.loadCats(coco.getCatIds())
nms=[cat['name'] for cat in cats]
print('COCO categories: \n{}\n'.format(' '.join(nms)))

nms = set([cat['supercategory'] for cat in cats])
print('COCO supercategories: \n{}'.format(' '.join(nms)))
```

```python
catIds = coco.getCatIds(catNms=['TYPE_VECHICLE','TYPE_PEDESTRIAN','TYPE_CYCLIST']);
imgIds = coco.getImgIds(catIds=catIds );
imgIds = coco.getImgIds(imgIds = [220])
img = coco.loadImgs(imgIds[np.random.randint(0,len(imgIds))])[0]
```

```python
import os
I = cv2.imread(os.path.join('/root/waymo/data_2d_od/train_small/',img['file_name']))
img_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
annIds = coco.getAnnIds(imgIds=img['id'], catIds=catIds, iscrowd=None)
anns = coco.loadAnns(annIds)
color_p = (255, 0, 0)
color_v = (0,255,0)
for annotation in anns:
    bbox = annotation['bbox']
    if annotation['category_id']==2:
        color = color_v
    else:
        color = color_p
    cv2.rectangle(I,(int(bbox[0]),int(bbox[1])),(int(bbox[0]+bbox[2]),int(bbox[1]+bbox[3])),color=color,thickness=2)
cv2.imwrite('temp.jpg',I)
```
