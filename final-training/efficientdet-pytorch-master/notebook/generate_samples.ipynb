```python
import os
for i in range(2,len(imgIds),100):
    imId = imgIds[i]
    print(imId)
    I = cv2.imread(os.path.join('/root/waymo/proper_split_dataset/valid/','{}.jpg'.format(imId)))
    img_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
    annIds = coco.getAnnIds(imId)
    anns = coco.loadAnns(annIds)
    color_p = (255, 0, 0)
    color_v = (0,255,0)
    color_b = (0,0,0)
    color_c = (0,0,255)
    for annotation in anns:
        bbox = annotation['bbox']
        if annotation['category_id']==1:
            color = color_v
        if annotation['category_id']==2:
            color = color_p
        if annotation['category_id']==3:
            color = color_c
        cv2.rectangle(I,(int(bbox[0]),int(bbox[1])),(int(bbox[0]+bbox[2]),int(bbox[1]+bbox[3])),color=color,thickness=2)
    cv2.imwrite(os.path.join('split_valid_annotations','{}_{}.jpg'.format('valid',imId)),I)
```
